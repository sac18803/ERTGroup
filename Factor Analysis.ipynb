# Factor Analysis with Python

Factor analysis is a dimensionality reduction technique commonly used in statistics. This is an unsupervised machine learning technique.

## Purpose

to use the BioChemist dataset from the pydataset module and perform a FA that creates two components

# Modules

```python
import pandas as pd 
from pydataset import data 
from sklearn.decomposition import FactorAnalysis 
import matplotlib.pyplot as plt
import numpy as np
```

# Data Prep

```python
df = data('bioChemists') 
df=df.iloc[1:250] 
X=df[['art','kid5','phd','ment']]
df
```

<p>The first line creates our dataframe called "df" and is made up of the dataset bioChemist</p>
<p>The second line reduces the df to 250 rows. This is done for the visual that we will make. To take the whole dataset and graph it would make a giant blob of color that would be hard to interpret.</p>
<p>The last line pulls the variables we want to use for our analysis. The meaning of these variables can be found by typing data("bioChemists",show_doc=True)</p>

```python
X
```

# Model Development

```python
fact_2c=FactorAnalysis(n_components=2) 
X_factor=fact_2c.fit_transform(X)
X_factor
```

The first line tells Python how many factors we want. The second line takes this information along with or revised dataset X to create the actual factors that we want.

# Visualization

To make the visualization requires several steps. We want to identify how well the two components separate students who are married from students who are not married. First, we need to make a dictionary that can be used to convert the single or married status to a number.

```python
thisdict = { "Single": "0", "Married": "1",}
thisdict
```

```python
z = np.array(df.mar.map(thisdict),dtype=np.int8)
colors = np.array(["yellow", "purple"])
z
```

Now the plot

```python
plt.scatter(X_factor[:,0],X_factor[:,1],c=colors[z])
```

<p>You can perhaps tell why we created the dictionary now. By mapping the dictionary to the mar variable it automatically changed every single and married entry in the df dataset to a 0 or 1. The c argument needs a number in order to set a color and this is what the dictionary was able to supply it with.</p>

<p>You can see that two factors do not do a good job of separating the people by their marital status. Additional factors may be useful but after two factors it becomes impossible to visualize them.</p>
